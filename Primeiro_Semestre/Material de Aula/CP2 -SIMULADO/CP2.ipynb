```python
# 1) Função para solicitar ao usuário o período a ser analisado
# 1) Uma função `solicitar_periodo` que solicita ao usuário o
# período: quantas semanas e quantos dias por semana.
# A função retorna uma lista com os dois valores.
 
def solicitar_periodo():
    semana = int(input("Digite quantas semanas deseja: "))
    dia = int(input("Digite quantos dias período: "))

    periodo = [semana, dia]
    return periodo


# 2) Função para registrar/mapear os consumos
# Uma função mapear_consumo para registrar/mapear os consumos.
# A função **`recebe`** uma lista com com os periodo - OKAY
# e ***retorna*** uma matriz preenchida pelos registros de consumos.

def mapear_consumo(lista_periodo):
    semanas = lista_periodo[0]
    dias = lista_periodo[1]
    matriz_periodo = []
    for s in range(semanas):
        matriz_periodo.append([])
        for d in range(dias):
            consumo = float(input(f"Digite o consumo do dia {d+1} da semana {s+1}: "))
            matriz_periodo[s].append(consumo)
        
    return matriz_periodo 


# 3) Função para identificar o maior e o menor consumo
# Uma função max_min_consumo para identificar o maior e o menor consumo no período mapeado.
# A função ***recebe**** uma matriz de consumo e
# ***retorna**** uma tupla com o valor maximo e minimo nesta ordem.

def max_min_consumo(max_min_consumo_periodo):
                                #semana e dia
    maximo = max_min_consumo_periodo[0][0]
    minimo = max_min_consumo_periodo[0][0]

    for s in range(len(max_min_consumo_periodo)): #acessar o indicie das semanas 
        for d in range(len(max_min_consumo_periodo[s])): #acessar o indicie dos dias
            
            if max_min_consumo_periodo[s][d] > maximo:
                maximo = max_min_consumo_periodo[s][d]
            
            if max_min_consumo_periodo[s][d] < minimo:
                minimo = max_min_consumo_periodo[s][d]
            
    return maximo, minimo

# 4) Função para listar os valores em que o consumo esteve abaixo da média
# Uma função valores_abaixo_media para listar os dias em que o consumo esteve abaixo da média.
# A função recebe 2 argumentos como parametros: uma matriz de consumo e a média do consumo e
# retorna uma lista apenas com os valores abaixo da média de consumo.

def valores_abaixo_media(consumo, media):
    media_baixa = []

    for c in range(len(consumo)):
        for m in range(len(consumo[c])):
            
            if consumo[c][m] < media:
                media_baixa.append(consumo[c][m])
    return media_baixa
                               
# 5) Função para calcular o consumo médio de água no período
# Uma função consumo_medio para calcular o consumo médio de água no período.
# A função recebe uma matriz de consumo e retorna a média do consumo.

def consumo_media(consumo_medio):
    soma = 0
    qtd = 0

    for c in range(len(consumo_medio)):
        for m in range(len(consumo_medio[c])):
            
            soma += consumo_medio[c][m]
            qtd += 1
            
    media = soma / qtd
    return media

# 6) Função principal para executar o programa

def main():
   periodo = solicitar_periodo()
   mapa_consumo = mapear_consumo(periodo)
   max_min_consumo_periodo = max_min_consumo(mapa_consumo)
   media = consumo_media(mapa_consumo)
   media_baixa_consumo = valores_abaixo_media(mapa_consumo, media)
   print ("Consumo de agora")
   print (f"Consumo maximo e mínimo {max_min_consumo_periodo}")
   print (f"Dias com consumo abaixo da média {media_baixa_consumo}")
   print (f"Consumo médio do período {media}")

   
 
# Não apagar as linhas abaixo.
if __name__ == '__main__':
    main()
```

```python
import csv

# Lista para armazenar registros
registros = []

def adicionar_registro(nome, idade, cidade):
    """Adiciona um novo registro à lista de registros."""
    registro = [nome, idade, cidade]
    registros.append(registro)
    print(f"Registro adicionado: {registro}")

def visualizar_registros():
    """Exibe todos os registros armazenados."""
    if registros:
        print("Registros:")
        for i, registro in enumerate(registros, start=1):
            print(f"{i}: Nome: {registro[0]}, Idade: {registro[1]}, Cidade: {registro[2]}")
    else:
        print("Nenhum registro encontrado.")

def salvar_registros_csv(nome_arquivo):
    """Salva os registros em um arquivo CSV."""
    try:
        with open(nome_arquivo, 'w', newline='') as arquivo:
            escritor = csv.writer(arquivo)
            # Escreve o cabeçalho
            escritor.writerow(["Nome", "Idade", "Cidade"])
            # Escreve cada registro
            for registro in registros:
                escritor.writerow(registro)
        print(f"Registros salvos com sucesso em '{nome_arquivo}'.")
    except Exception as e:
        print(f"Ocorreu um erro ao salvar os registros: {e}")

def menu():
    """Exibe o menu principal e processa as opções do usuário."""
    while True:
        print("\nMenu:")
        print("1. Adicionar registro")
        print("2. Visualizar registros")
        print("3. Salvar registros em CSV")
        print("4. Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            nome = input("Digite o nome: ")
            idade = input("Digite a idade: ")
            cidade = input("Digite a cidade: ")
            adicionar_registro(nome, idade, cidade)
        elif opcao == "2":
            visualizar_registros()
        elif opcao == "3":
            nome_arquivo = input("Digite o nome do arquivo CSV para salvar: ")
            salvar_registros_csv(nome_arquivo)
        elif opcao == "4":
            print("Saindo do programa.")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Executa o menu
menu()
```
